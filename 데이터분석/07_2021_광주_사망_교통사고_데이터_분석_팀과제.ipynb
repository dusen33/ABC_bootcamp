```python
# 필요한 라이브러리 임포트
import pandas as pd

import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib

import folium
```

```python
# 데이터 불러오기
df = pd.read_csv('/content/도로교통공단_사망 교통사고 정보_20211231.csv', encoding='EUC-KR')
df
```

```python
# 데이터 전처리 전
df.info()
```

```python
# 발생시간 칼럼 생성
df['발생시간'] = df['발생년월일시'].str[11:13]
df.head()
```

```python
# 발생시간 int 타입으로 변경
df = df.astype({'발생시간' : 'int'})
df.info()
```

```python
# 발생년월일 칼럼 생성
df['발생년월일'] = df['발생년월일시'].str[:11]
df.head()
```

```python
# 발생년월일시 칼럼 삭제
del df['발생년월일시']
df.head()
```

```python
# 칼럼 순서 정리
col1 = df.columns[-2:].to_list()
col1.reverse()
col2 = df.columns[1:-2].to_list()
col_list = col1+col2
col_list.insert(0, '발생년')

df = df[col_list]
```

```python
# 칼럼 순서 잘 바뀌었나 확인
df.columns
```

```python
# 데이터 전처리 후
df.info()
```

```python
# 정제된 데이터셋 저장
df.to_csv('/content/2021_도로교통공단_교통사고 데이터 전처리.csv', encoding='utf-8-sig', index=False)
```

```python
# 광주 교통사고 데이터 분석
dj_df = df[df['발생지시도']=='광주']
dj_df.info()
```

```python
# 광주의 발생시간별 교통사고 사망자 현황
fig = px.bar(dj_df, x = '발생시간', y = '사망자수')
fig.show()
```

```python
# 광주의 지역구별 교통사고 사망자 현황
fig = px.bar(dj_df, x = '사망자수', y = '발생지시군구', orientation='h')
fig.show()
```

```python
# 광주의 요일별 교통사고 사망자 현황
fig = px.bar(dj_df, x = '사망자수', y = '요일', orientation='h', category_orders={'요일':['월', '화', '수', '목', '금', '토', '일']})
fig.show()
```

```python
# 광주 지도 준비하기
import folium

map = folium.Map(location=[35.164553702943266, 126.8508425550919 ]) # 경도, 위도
map
```

```python
# 사망자수에 따른 마커 크기 시각화
dj_df = dj_df.astype({'사망자수': 'float', '부상자수':'float'})
dj_df.head()
```

```python
# 지도 준비 -> CIrcleMarker 표출
map = folium.Map(location=[35.164553702943266, 126.8508425550919 ])

for n in dj_df.index:
  cnt = dj_df['사망자수'][n] + dj_df['부상자수'][n]

  lat = dj_df['위도'][n]
  lng = dj_df['경도'][n]

  folium.CircleMarker([lat, lng], radius = cnt*10, popup = dj_df['사고유형'][n],
                      color='#CD0000', fill_color='#CD0000').add_to(map)

map
```

```python
# 시각화된 지도 저장
map.save('2021_광주교통사고현황.html')
```
