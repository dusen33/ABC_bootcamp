우리동네 연령별 인구수 result 리스트에 저장 -> 시각화를 위해

```python
import csv
import matplotlib.pyplot as plt


# 1) 연령별 인구수 리스트 선언
result = []


# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 데이터 한 줄씩 출력
for row in data:
  if '신도림동' in row [0]: # 읍면동 조건
    # print(row[3:]) # 0 ~ 100세 정보
    for i in row[3:]:
      result.append(int(i))

# 4) 파일 닫기
f.close()


# 5) 인구구조 시각화
plt.title('신도림동의 인구구조')
plt.style.use('ggplot') # 격자 무늬 스타일
plt.plot(result)
plt.show()
```

사용자 입력을 통한 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result = []

dong = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')

# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 연령별 인구수 리스트 담기
for row in data:
  if dong in row[0]: # 읍면동 조건
    print(row[3:]) # 0~100세 정보
    for i in row[3:]:
      result.append(int(i.replace(",", "")))
    break

# 4) 파일 닫기
f.close()

# 5) 인구구조 시각화
plt.title(dong + '의 인구구조')
plt.style.use('ggplot')

plt.plot(result)
plt.show()
```

두 개의 동네의 인구구조 비교하기

```python
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result1 = []
result2 = []

dong1 = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')
dong2 = input('인구구조를 확인하고 싶은 동네명(동)을 입력하세요: ')

# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3) 데이터 한 줄씩 출력
for row in data:
  if dong1 in row [0]: # 읍면동 조건
    for i in row[3:]:
      result1.append(int(i.replace(",", "")))


  if dong2 in row [0]: # 읍면동 조건
    for i in row[3:]:
      result2.append(int(i.replace(",", "")))


# 4) 파일 닫기
f.close()


# 5) 인구구조 시각화
plt.title(dong1 + ', ' + dong2 + '의 인구구조')
plt.style.use('ggplot') # 격자 무늬 스타일

plt.plot(result1, color = 'r', label = dong1)
plt.plot(result2, color = 'b', label = dong2)
plt.legend()

plt.show()
```

```python
# 하나의 for문에서 두개의 동네 인구수를 담으세요.
import csv
import matplotlib.pyplot as plt

# 1) 연령별 인구수 리스트 선언
result_1 = []
result_2 = []
address_1 = input('비교할 동네 전체 주소(시군구 읍면동) 입력하세요 : ')
address_2 = input('우리 동네 전체 주소(시군구 읍면동) 입력하세요 : ')

# 2) 파일 열기
f = open('/content/201902_201902_연령별인구현황_월간.csv')
data = csv.reader(f)

# 3 ) 연령별 인구수 리스트 담기
for row in data :
  split_address = row[0].split('(')[0].strip() # strip() -> 앞 뒤 공백 제거
  if split_address == address_1: # 읍면동 조건
    for i in row[3:] :
      result_1.append(int(i.replace(",","")))

  if split_address == address_2: # 읍면동 조건
    for i in row[3:] :
      result_2.append(int(i.replace(",","")))

# 4) 파일을 닫기.
f.close()

# 5) 인구구조 시각화
plt.title('%s와 %s의 인구구조 비교' %(address_1, address_2))
plt.style.use('ggplot')
plt.plot(result_1, 'r', label= address_1)
plt.plot(result_2, 'b', label=address_2)
plt.legend()
plt.show
```
